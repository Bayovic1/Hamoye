```python
import pandas as pd
df_food = pd.read_csv("foodbalancesheets_e_africa_noflag - Sheet1 (1).csv")
df_food
```

```python
#What is the total Protein supply quantity in Madagascar in 2015?

grp_food = df_food.groupby("Area").get_group("Madagascar") #Selecting Madagascar from the chunks of Areas
print("List of Madagascar are :", grp_food)
sup_prot = grp_food.groupby("Element")["Y2015"].sum() #selecting the column of interest which is the Y2015 and grouping with Element which contains the protein supplied
print("")
print("Element sum list for Y2015 :")
sup_prot
```

```python
# Select columns ‘Y2017’ and ‘Area’, Perform a groupby operation on ‘Area’.  Which of these Areas had the highest sum in 2017?

YH2017 = df_food.groupby("Area")["Y2017"].sum()
print("The Highest is :", YH2017.max()) #print the Highest Value
print(YH2017 == 1483268.23 ) #returns only the True Value
print("The Country with the highest is :", YH2017[YH2017 == 1483268.23] ) # Display only the True Value
```

```python

#What is the mean and standard deviation across the whole dataset for the year 2017 to 2 decimal places?
yt = df_food.describe()
yt
```

```python
print("Standard Deviation for 2017 is :", round(df_food['Y2017'].std(),2))
print("Mean for 2017 is :", round(df_food['Y2017'].mean(),2))
```

```python

#Perform a groupby operation on ‘Element’.  What is the total number of the sum of Processing in 2017?
pro2017 = df_food.groupby("Element").get_group("Processing")
print("List of all the Processing Element are ;", pro2017)
print("Sum of the Processing for the Year 2017 is :", pro2017["Y2017"].sum())
```

```python
#Select columns ‘Y2017’ and ‘Area’, Perform a groupby operation on ‘Area’.  Which of these Areas had the 7th lowest sum in 2017?
Low_2017 = df_food.groupby("Area")["Y2017"].sum()
L27 = Low_2017.sort_values(ascending=True) #sorting the values accordingly
L27.head(10) #listing the sorted values accordingly
```
